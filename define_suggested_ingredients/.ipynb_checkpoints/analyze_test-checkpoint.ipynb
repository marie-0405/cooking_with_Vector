```python
#  材料を集計する
%matplotlib inline 
import matplotlib.pyplot as plt
my_recipe = None
# 取得したrecipe_idのデータを取得
for i in range(len(recipe_id)):
    tmp = tea.loc[tea['recipe_id'] == recipe_id[i]]
    my_recipe = pd.concat([my_recipe, tmp])
my_recipe = my_recipe.reset_index(drop=True)
analyze_ingredient = {}
# print(type(my_recipe['tsukurepo_count'][0]))
plt.rcParams["font.family"] = "MS Gothic"
# 辞書型で材料とつくれぽ値を取得
# print(my_recipe['ingredient'][i] in analyze_ingredient.keys())
for i in range(len(my_recipe)):
    if my_recipe['ingredient'][i] in analyze_ingredient.keys():
        # もし材料名が既に存在していたら、つくれぽ数を足す
        analyze_ingredient[my_recipe['ingredient'][i]] += my_recipe['tsukurepo_count'][i]
#         analyze_ingredient[my_recipe['ingredient'][i]] += 1

    else:
        analyze_ingredient[my_recipe['ingredient'][i]] = my_recipe['tsukurepo_count'][i]
#         analyze_ingredient[my_recipe['ingredient'][i]] = 1
analyze_ingredient = sorted(analyze_ingredient.items(), key=lambda x: x[1], reverse=True)
# analyze_ingredient = list(analyze_ingredient)

# 80%までは、その値をとり、それ以外は”その他”に分類する

# 合計値summを求める
summ = 0
for i in range(len(analyze_ingredient)):
    summ += analyze_ingredient[i][1]
flg = 0
tmp = 0
for i in range(len(analyze_ingredient)):
    tmp += analyze_ingredient[i][1]
    if tmp >= summ*0.8:
        flg = analyze_ingredient[i][0]
        flg = i + 1
        break
others = 0  # その他のカウント数
for i in range(flg, len(analyze_ingredient)):
    others += analyze_ingredient[i][1]
    
# その他に分類されるデータを削除
for i in range(len(analyze_ingredient) - flg):
    del analyze_ingredient[flg]

# 新しくその他を分類
analyze_ingredient.append(('その他', others)) 

label = []
data = []
label = [analyze_ingredient[i][0] for i in range(len(analyze_ingredient))]
data = [analyze_ingredient[i][1] for i in range(len(analyze_ingredient))]
# print(label)
# print(data)
plt.pie(data, labels=label, startangle=90, autopct="%1.1f%%")  #  autopct="%1.1f%%"
plt.axis('equal')
```
